# Conditional statements (if, elif, else).

```python
number = 10

if 25 < number:
    print("25 is greater than 10")
```

```python
number = 10

if 25 < number:
    print("25 is greater than 10")
elif 25 > number:
    print("25 is greater than 10")
```

```python
number = 10

if 25 > number:
    print("Yes")
else:
    print("No")
```

# Logical expressions in conditions.

```python
age = 34

if age < 18:
    print("Young")
elif age > 18 and age < 60:
    print("Adult")
else:
    print("Senior citizen")
```

```python
temp = 25

if temp < 20 or temp > 30:
    print("uncomfortable")
else:
    print("comfortable")
```

```python
grad = "No"

if grad != "yes":
    print("ineligible")
else:
    print("eligible")
```

```python
list_num = [1,2,3,4,5,6,7,8,9]

list_even_num = []

for num in list_num:
    if num % 2 == 0:
        list_even_num.append(num)
```
